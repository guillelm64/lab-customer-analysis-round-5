```python
#Do the distributions for different numerical variables look like a normal distribution
#def no_out_liers(dt_, atrib):
def log_transfom_clean1(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return 0

nume_log = nume['income'].apply(lambda x :log_transfom_clean1(x))

def log_transfom_clean2(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN 

nume['nume_log'] = list(map(log_transfom_clean2, nume['total claim amount']))
nume['nume_log'] = nume['nume_log'].fillna(np.mean(nume['nume_log']))
sns.displot(nume_log)
plt.show()


nume_log = nume['total claim amount'].apply(lambda x :log_transfom_clean1(x))
nume['nume_log'] = list(map(log_transfom_clean2, nume['total claim amount']))
nume['nume_log'] = nume['nume_log'].fillna(np.mean(nume['nume_log']))
sns.displot(nume_log)
plt.show()
```
